```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Literal
from pydantic import BaseModel, Field
from langchain_ollama import ChatOllama
```

```python
llm = ChatOllama(
    model='llama3.2:latest'
)
```

```python
class Sample(BaseModel):
    review : str = Field(description="enter the review")
    sentiment : Literal['positive','negative'] = Field(description="nothing to say")
```

```python
# result = llm.with_stuctured_output(Sample).invoke("the food was very good")
```

```python
from pydantic import BaseModel, Field
from langchain.agents import create_agent

agent = create_agent(
    model=llm,
    response_format=Sample  
)

result = agent.invoke({
    "messages": [{"role": "user", "content": "the food was very bad"}]
})

print(result)
# ContactInfo(name='John Doe', email='john@example.com', phone='(555) 123-4567')
```
